__Q9. Advanced Filtering + Subsetting__

**Context: Training data is often filtered based on business rules.**

Dataset: Tips

Tasks:

1. Filter: total_bill > 20 AND tip < 3
2. Select only relevant columns
3. Analyze: Is this segment under-tipping?
4. Explain: How such filtering helps anomaly detection

```python
#1. Filter: total_bill > 20 AND tip < 3
import seaborn as sns
import pandas as pd

tipsDataSet = sns.load_dataset("tips")
#tipsDataSet.head(5)

filteredData = tipsDataSet[(tipsDataSet["total_bill"] > 20) & (tipsDataSet["tip"] < 3)]
print(filteredData)
```

```python
#2. Select only relevant columns
tipsDataSet = sns.load_dataset("tips")
#tipsDataSet.head(5)

filteredData = tipsDataSet[(tipsDataSet["total_bill"] > 20) & (tipsDataSet["tip"]  < 3)]
releventColumns = filteredData[["total_bill", "tip", "day"]]
print(releventColumns)
```

```python
#3. Analyze: Is this segment under-tipping?
yes it is under tipping as the tip is relatively lower with respect to the spending (total_bill)
```

```python
#4. Explain: How such filtering helps anomaly detection
Filtering helps anomaly detection by defining explicit conditions that capture deviations from expected behavior.
```
