```python
#2. Group by: day and time
import seaborn as sns
import pandas as pd

# Load tips dataset
df = sns.load_dataset("tips")

# Create tip_percentage
df["tip_percentage"] = (df["tip"] / df["total_bill"]) * 100

# Group by day and time and compute average tip percentage
grouped_tips = df.groupby(["day", "time"],observed=True)["tip_percentage"].mean().reset_index().dropna()

print(grouped_tips)
```

```python
#3. Identify: Highest revenue segment

import seaborn as sns
import pandas as pd

# Load tips dataset
df = sns.load_dataset("tips")

# Group by day and time and compute total revenue
revenue_by_segment = (
    df.groupby(["day", "time"])["total_bill"]
      .sum()
      .reset_index()
)

# Display revenue per segment
print(revenue_by_segment)

# Identify the highest revenue segment
highest_revenue_segment = revenue_by_segment.loc[
    revenue_by_segment["total_bill"].idxmax()
]

print("\nHighest revenue segment:")
print(highest_revenue_segment)

```
