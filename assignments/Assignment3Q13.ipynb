**Q13. Custom Feature Transformation using apply()**

Context: Real ML requires custom logic transformation.

Dataset: Titanic

Tasks:

1. Create age_group: Child / Adult / Senior
2. Use .apply()
3. Compute survival rate per group
4. Interpret: Which segment has highest survival likelihood

```python
#1. Create age_group: Child / Adult / Senior

import pandas as pd

df = pd.read_csv("titanic.csv")

#df.columns = df.columns.str.lower()
df["age"] = df["age"].fillna(df["age"].median())

df["age_group"] = pd.cut(
    df["age"],
    bins=[0, 18, 60, df["age"].max()],
    labels=["Child", "Adult", "Senior"],
    right=False
)

df[["age", "age_group"]].head(10)
```

```python
#2 Use .apply()

import pandas as pd

df = pd.read_csv("titanic.csv")
df.columns = df.columns.str.lower()
df["age"] = df["age"].fillna(df["age"].median())
df["age_group"] = df["age"].apply(lambda x: "Child" if x < 18 else ("Adult" if x < 60 else "Senior"))
df[["age", "age_group"]].head(10)
```

```python
#3. Compute survival rate per group

import pandas as pd

# Load dataset
df = pd.read_csv("titanic.csv")

# Normalize column names (good practice)
df.columns = df.columns.str.lower()

# Handle missing age values
df["age"] = df["age"].fillna(df["age"].median())

# Create age_group using pd.cut (from previous task)
df["age_group"] = pd.cut(
    df["age"],
    bins=[0, 18, 60, df["age"].max()],
    labels=["Child", "Adult", "Senior"],
    right=False
)

# Compute survival rate per age group
survivalRate = df.groupby("age_group")["survived"].mean()
print(survivalRate)
```

```python
#4. Interpret: Which segment has highest survival likelihood


print("Children had the highest survival",df.groupby("age_group")["survived"].mean())
```
