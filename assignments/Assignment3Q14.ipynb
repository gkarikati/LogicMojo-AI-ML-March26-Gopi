```python
#2. Compute: Mean Standard deviation
import numpy as np

# Set seed for reproducibility (optional but recommended)
np.random.seed(42)

# Generate 1000 random error values (mean=0, std=1)
errors = np.random.normal(loc=0, scale=1, size=1000)
# errors = np.random.normal(loc=0, scale=1, size=1000)

# Compute mean and standard deviation
error_mean = np.mean(errors)
error_std = np.std(errors)

print("Mean of errors:", error_mean)
print("Standard deviation of errors:", error_std)
```

```python
#3. Identify outliers: Using mean ± 2*std

# Identify outliers using mean ± 2*std
lower_limit = error_mean - 2 * error_std
upper_limit = error_mean + 2 * error_std

outliers = errors[(errors < lower_limit) | (errors > upper_limit)]

print("Lower limit:", lower_limit)
print("Upper limit:", upper_limit)
print("Number of outliers:", len(outliers))
print("Sample outliers:", outliers[:10])

```

```python
#4. Explain: What high variance means in model performance

High variance indicates overfitting, where the model learns noise from the training data and fails to generalize well to unseen data.
```
