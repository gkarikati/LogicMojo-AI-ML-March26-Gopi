Q1. Understanding Feature Distributions (Iris Dataset)

Context: Before training any ML model, we analyze feature distributions.

Dataset: Load Iris dataset from sklearn

Tasks:

1. Extract feature matrix as NumPy array
2. Compute: Mean Median Standard deviation Variance (for each feature)
3. Identify: Which feature has highest variability and why it matters in ML
4. Convert any one feature into shape (n,1) and explain why ML models expect this format

```python
#1.Extract feature matrix as NumPy array

from sklearn.datasets import load_iris
import numpy as np

iris = load_iris() # Loading the Iris dataset
X = iris.data #feature matrix
#print(X) # Print 
print(X.shape) #feature matrix
print(type(X)) #Pring data type
```

```python
# 2. Compute: Mean Median Standard deviation Variance (for each feature)

# Feature matrix
X = iris.data
feature_names = iris.feature_names

print(feature_names)


mean = np.mean(X, axis=0)
median = np.median(X, axis=0)
std_dev = np.std(X, axis=0)
variance = np.var(X, axis=0)

# Display results
for i in range(len(feature_names)):
    print(f"{feature_names[i]}:")
    print(f"  Mean     = {mean[i]}")
    print(f"  Median   = {median[i]}")
    print(f"  Std Dev  = {std_dev[i]}")
    print(f"  Variance = {variance[i]}\n")
```
