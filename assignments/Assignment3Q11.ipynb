```python
#5.Convert to NumPy array
import pandas as pd
from sklearn.preprocessing import StandardScaler

#Load dataset
df = pd.read_csv("titanic.csv")

# Normalize column names
df.columns = df.columns.str.lower()

# Handle missing values
df["age"].fillna(df["age"].median(), inplace=True)
df["embarked"].fillna(df["embarked"].mode()[0], inplace=True)

#Encode categorical variables
df["sex"] = df["sex"].map({"male": 0, "female": 1})
df = pd.get_dummies(df, columns=["embarked"], drop_first=True)

#DROP remaining categorical columns (IMPORTANT)
cat_cols_to_drop = ["class", "who", "alive", "embark_town", "deck"]
df.drop(columns=[c for c in cat_cols_to_drop if c in df.columns], inplace=True)

#Normalize numerical features
num_cols = ["age", "fare", "sibsp", "parch", "pclass"]
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

#Convert to NumPy
X = df.drop(columns=["survived"])
y = df["survived"]

X_np = X.to_numpy().astype("float32")
y_np = y.to_numpy().astype("int32")

print(X_np.shape, y_np.shape)
```
