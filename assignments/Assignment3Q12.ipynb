```python
# Identify highly correlated feature pairs
import pandas as pd
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()

# Create DataFrame
df = pd.DataFrame(iris.data,columns=iris.feature_names)

corr_matrix = df.corr()

#print(corr_matrix)
#df.head()
corr_matrix


high_corr_pairs = []

threshold = 0.8

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        corr_value = corr_matrix.iloc[i, j]
        if abs(corr_value) >= threshold:
            high_corr_pairs.append((corr_matrix.columns[i],corr_matrix.columns[j],corr_value))

high_corr_pairs




```

```python
#3. Drop redundant features

df_reduced = df.drop(columns=["petal width (cm)"])

df_reduced.head()
```

```python
#4. Explain: Impact on overfitting and model performance
Dropping redundant, highly correlated features reduces overfitting by lowering model complexity and variance, 
leading to better generalization and more stable performance on unseen data.
```
