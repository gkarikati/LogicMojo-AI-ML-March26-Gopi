**Q8. Group-Based Insights (Customer Segmentation Logic)**


Context: ML models often depend on aggregated patterns.

Dataset: Tips

Tasks:

1. Compute: Average tip by day Total bill by gender
2. Create new feature: tip_percentage
3. Identify: Which day has highest tipping behavior
4. Explain: How this helps in segmentation models

```python
#1. Compute: Average tip by day Total bill by gender
import seaborn as sns
import pandas as pd
tipsDataSet = sns.load_dataset("tips")
#tipsDataSet.head(5)
print("********Average tip by day**********")
avgTipByDay = tipsDataSet.groupby("day",observed=True )["tip"].mean()

print(avgTipByDay)
print("********Total bill by gender**********")
totalBillByGender = tipsDataSet.groupby("sex",observed=True )["total_bill"].sum()

print(totalBillByGender)
```

```python
#2. Create new feature: tip_percentage
#tipsDataSet.head(5)
tipPercentage = ((tipsDataSet["tip"]/tipsDataSet["total_bill"])*100)
tipsDataSet["tips_percentage"] = (tipPercentage)
tipsDataSet.head(5)
```

```python
#3. Identify: Which day has highest tipping behavior
avgTipsByDay = tipsDataSet.groupby("day",observed=True)["tip"].mean()
highestTipsOfDay = avgTipsByDay.idxmax()
print(highestTipsOfDay)
#tipsDataSet.head(5)
```

```python
#4. Explain: How this helps in segmentation models
Segmentation helps ML models learn different behaviors for different groups instead of treating all data the same.
It will improve accuracy in the prediction.
```
