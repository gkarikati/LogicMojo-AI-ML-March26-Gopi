**Q5. Filtering Data for Business Logic**


Context: Real ML problems require filtering relevant populations.

Dataset: Titanic

Tasks:

1. Filter: Female passengers in 1st class
2. Compute: Survival rate for this group
3. Compare with overall survival rate
4. Interpret: What insight can be used in ML feature engineering

```python
#1.Filter: Female passengers in 1st class
import seaborn as sbn
import pandas as pd

titanicDataSet = sbn.load_dataset("titanic")
#print(titanicDataSet)
first_class_females = titanicDataSet[(titanicDataSet["sex"] == "female") &
    (titanicDataSet["pclass"] == 1)]

print(first_class_females.shape[0])
print(first_class_females)


 
    
```

```python
# 2 Compute: Survival rate for this group


survival_rate = first_class_females["survived"].mean()

print("Survival rate:", survival_rate)

```

```python
#3 Compare with overall survival rate


first_class_females = titanicDataSet[(titanicDataSet["sex"] == "female") & (titanicDataSet["pclass"] == 1)]

group_survival_rate = first_class_females["survived"].mean()
overall_survival_rate = titanicDataSet["survived"].mean()

print("Survival rate of Female from 1st class):", group_survival_rate)
print("Overall survival rate:", overall_survival_rate)
```

```python
#4. Interpret: What insight can be used in ML feature engineering

From the Titanic analysis, we observed that female passengers in first class had a survival rate much higher than the overall population.
This indicates a strong interaction between gender and passenger class. In feature engineering, this insight can be used to create interaction 
features that explicitly capture such combinations, which helps models learn important patterns more effectively, especially linear models
```
