Context: Models are trained on selective features, not full raw data.

Dataset: Iris

Tasks:

1. Extract: First 100 samples Only last 2 features
2. Use boolean masking: Select samples where petal length is greater than dataset mean
3. Count selected samples

```python
#Task 1 Extract: First 100 samples Only last 2 features
from sklearn.datasets import load_iris
import numpy as np
irisData = load_iris()
#print(irisData)
ir = irisData.data
#print(ir) # without the headers
petal_length = ir[:100,2:]
print(petal_length.shape)
print(petal_length)

```

```python
#2: Use boolean masking: Select samples where petal length is greater than dataset mean
from sklearn.datasets import load_iris
import numpy as np
irisData = load_iris()
ird = irisData.data # this is full data with 150 rows and 4 columns (features)
#print(ird)
petal_length = ird[:,2]
#print(petal_length)
petal_length_mean = np.mean(petal_length)
#print(petal_length_mean)
petal_mask = petal_length > petal_length_mean
#print(petal_mask)
# mask on selected selected petals only
ird_selected_for_mask = ird[petal_mask]
print("Mean petal length:",petal_length_mean)
print("selected sample's Shape",ird_selected_for_mask.shape)


```

```python
#task 3: Count selected samples
print("Count of selected samples:", ird_selected_for_mask.shape[0])
```
