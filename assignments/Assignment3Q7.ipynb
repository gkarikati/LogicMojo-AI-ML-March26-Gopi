**Q7. Feature Engineering for Model Improvement**

Context: Raw data is rarely useful—features must be created.

Dataset: Titanic

Tasks:

1. Create: family_size = sibsp + parch + 1
2. Create: is_alone (binary feature)
3. Compute survival rate by is_alone   
4. Explain: Why engineered features improve ML models

```python
#1. Create: family_size = sibsp + parch + 1
import seaborn as sns
import pandas as pd
titanicDataSet = sns.load_dataset("titanic")
#titanicDataSet.head(10)
# New feature family_size
titanicDataSet["family_size"] = (titanicDataSet["sibsp"]+titanicDataSet["parch"] + 1)
titanicDataSet.head(5)
```

```python
#2. Create: is_alone (binary feature)
titanicDataSet["family_size"] = (titanicDataSet["sibsp"]+titanicDataSet["parch"] + 1)
titanicDataSet["is_alone"] = (titanicDataSet["family_size"] == 1)
titanicDataSet.head(5)

```

```python
#3. Compute survival rate by is_alone
titanicDataSet["family_size"] = (titanicDataSet["sibsp"]+titanicDataSet["parch"] + 1)
titanicDataSet["is_alone"] = (titanicDataSet["family_size"] == 1)

survival_rate_by_is_alone = (titanicDataSet.groupby("is_alone")["survived"].mean())
print(survival_rate_by_is_alone)
```

```python
# 4. Explain: Why engineered features improve ML models
The purpose of creating feature engineering fields is to improve the machine learning models as they convert the raw
data to meaningful data or patters which is easy for model to learn or improve accuracy.
```
