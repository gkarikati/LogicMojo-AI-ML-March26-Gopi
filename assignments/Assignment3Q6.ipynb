**Q6. Handling Missing Data (Critical ML Step)**


Context: Models cannot handle null values.

Dataset: Titanic

Tasks:

1. Fill missing age using median
2. Fill missing embarked using mode
3. Drop deck column
4. Explain: Why different strategies are used for different columns

```python
# 1. Fill missing age using median
import seaborn as sns
import pandas as pd
titanicDataSet = sns.load_dataset("titanic")
median_age = titanicDataSet["age"].median()    
#print(median_age)
print(titanicDataSet["age"].isnull().sum())
titanicDataSet["age"].fillna(median_age, inplace=True)
print(titanicDataSet["age"].isnull().sum())
```

```python
#2. Fill missing embarked using mode
#print(titanicDataSet["embarked"])
print(titanicDataSet["embarked"].isnull().sum())


embarked_mode = titanicDataSet["embarked"].mode()[0]
print(embarked_mode)
titanicDataSet["embarked"] = titanicDataSet["embarked"].fillna(embarked_mode)
print(titanicDataSet["embarked"].isnull().sum())

```

```python
#3. Drop deck column
import seaborn as sns
import pandas as pd
titanicDataSet = sns.load_dataset("titanic")
#print(titanicDataSet)
titanicDataSet = titanicDataSet.drop(columns=["deck"])
print("deck" in titanicDataSet.columns)
```

```python
#Explain: Why different strategies are used for different columns
Different columns represent different data types quality, so they require different preprocessing strategies. 
Applying the same treatment to all columns can stain data, or even break machine learning models.
```
